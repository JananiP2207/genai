```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.google.com/about/careers/applications/jobs/results")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain.prompts import ChatPromptTemplate

prompt_extract = ChatPromptTemplate.from_template("""
### SCRAPED TEXT FROM WEBSITE:
{page_data}

### INSTRUCTION:
The scraped text is from a JOB SEARCH RESULTS page.

Extract ALL job listings and return a JSON ARRAY.
Each job object must contain:
- role
- company
- location
- experience_level
- short_description (if present, else null)

Rules:
- Do NOT invent missing information
- If a field is not clearly present, use null
- Return ONLY valid JSON
- No explanations, no markdown
""")

chain_extract = prompt_extract | llm
res = chain_extract.invoke({"page_data": page_data})

print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
from langchain.prompts import PromptTemplate

prompt_email = PromptTemplate.from_template(
    """
### JOB DESCRIPTION:
{job_description}

### INSTRUCTION:
You are Mohan, a Business Development Executive at AtliQ.

AtliQ is an AI & Software Consulting company that helps enterprises with:
- AI-driven automation
- Scalable software systems
- Process optimization
- Cost reduction

Write a professional cold email to the company based on the job above,
explaining how AtliQ can support their requirements.

Include only the MOST relevant portfolio links from below:
{link_list}

Constraints:
- No preamble
- No subject line
- Professional tone
- Clear value proposition

### EMAIL:
"""
)
```

```python
job_description = f"""
Role: {job.get('role')}
Company: {job.get('company')}
Location: {job.get('location')}
Experience Level: {job.get('experience_level')}
Description: {job.get('short_description')}
"""
```

```python
link_list = "\n".join(
    [
        f"- {item['links']}" if isinstance(item, dict)
        else f"- {item[0]}" if isinstance(item, list)
        else f"- {item}"
        for item in links
    ]
)
print(links)
print(type(links[0]))
```
